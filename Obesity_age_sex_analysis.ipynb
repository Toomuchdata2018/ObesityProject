```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read csv files
age_sex_2012 = pd.read_csv("ObesityAgeAndSex_2012.csv")
age_sex_2013 = pd.read_csv("ObesityAgeAndSex_2013.csv")
age_sex_2014 = pd.read_csv("ObesityAgeAndSex_2014.csv")
age_sex_2015 = pd.read_csv("ObesityAgeAndSex_2015.csv")
age_sex_2016 = pd.read_csv("ObesityAgeAndSex_2016.csv")
```

```python
# age_sex_2016.head()
```

```python
age_sex_2016["BMI Category"] = age_sex_2016["BMI Category"].replace({"Normal Weight":"Others","Don't know/Refused/Missing":"Others",
                                                                     "Underweight":"Others","Overweight":"Others"})
# age_sex_2016.head()
```

```python
grouped_age_sex_2016 = age_sex_2016.groupby(["Age Group","Sex","BMI Category"])["Date"].count()
# grouped_age_sex_2016
```

```python
# Unstacking the BMI category
grouped_age_sex_2016_df = grouped_age_sex_2016.unstack(level=2)
# grouped_age_sex_2016_df
```

```python
# Changing all BMI categories, except "Obesity", to "Others"
grouped_age_sex_2016_df["Total Count"] = grouped_age_sex_2016_df["Obese"] + grouped_age_sex_2016_df["Others"]
# grouped_age_sex_2016_df
```

```python
# Adding a new column named: "Total% by Sex by Age Group" to capture percentages
grouped_age_sex_2016_df["Total% by Sex by Age Group"] = (grouped_age_sex_2016_df["Obese"] / grouped_age_sex_2016_df["Total Count"]) * 100
# grouped_age_sex_2016_df
```

```python
# Resetting index
grouped_age_sex_2016_df.reset_index(drop=False, inplace=True, col_fill='')
# grouped_age_sex_2016_df
```

```python
# Removing rows that contain "Refused" in order to only show "Male" and "Female" categories in Sex
grouped_age_sex_2016_df = grouped_age_sex_2016_df[grouped_age_sex_2016_df.Sex.str.contains("Refused") == False]
# grouped_age_sex_2016_df
```

```python
# Adding a new column named "Year"
grouped_age_sex_2016_df["Year"] = ""
grouped_age_sex_2016_df["Year"] = grouped_age_sex_2016_df["Year"].replace({"":"2016"})
# grouped_age_sex_2016_df
```

```python
# Setting Year as index to move column to first position
grouped_age_sex_2016_df = grouped_age_sex_2016_df.set_index(["Year"],drop=True,inplace=False)
# grouped_age_sex_2016_df
```

```python
# Resetting index
grouped_age_sex_2016_df.reset_index(drop=False, inplace=True)
```

```python
# Formatting
pd.set_option("display.float_format", "{:,.0f}".format)
# grouped_age_sex_2016_df
```

```python
grouped_age_sex_2016_df.style.format({
    'Total% by Sex by Age Group': '{:,.2f}%'.format,})
```

```python
# Plotting a bar chart to see % of obesity in different Age groups
# sns.set()
# sns.factorplot(x="Age Group", y='Total% by Sex by Age Group', hue='Sex', data=grouped_age_sex_2016_df, kind='bar',size=7)
# plt.show()
```

```python
# Plotting a line chart to see % of obesity in different Age groups

sns.set()

sns.factorplot(x="Age Group", y='Total% by Sex by Age Group', hue='Sex', data=grouped_age_sex_2016_df, kind='point',size=7)
plt.title("2016",size=20)

plt.show()
```
